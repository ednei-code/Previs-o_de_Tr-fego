## Definição do Problema

A análise de dados sobre congestionamentos e padrões de tráfego é fundamental para a elaboração de soluções eficazes em planejamento urbano, infraestrutura e gestão do trânsito em áreas metropolitanas.

## Os Dados
Este conjunto de dados, alimentado por um modelo de visão computacional, oferece um retrato detalhado do tráfego. Ele contabiliza carros, bicicletas, ônibus e caminhões a cada 15 minutos, registrando a hora, data e dia da semana. Além do número total de veículos, o conjunto classifica o tráfego em quatro níveis (baixo, normal, alto e pesado), fornecendo uma visão completa da movimentação em um período de um mês.

## Possibilidades
Este conjunto de dados é uma ferramenta valiosa para melhorar o transporte urbano. Ao analisar os dados sobre o fluxo de veículos e os pontos de congestionamento, podemos identificar áreas que precisam de melhorias na infraestrutura, como semáforos mais eficientes ou novas faixas. Essa informação permite tomar decisões mais precisas sobre o planejamento de rotas, a construção de novas vias e a gestão do tráfego, tornando as cidades mais fluidas e menos congestionadas.

## Objetivo:

Este projeto tem como objetivo desenvolver um modelo de previsão de tráfego utilizando técnicas de machine learning. Através da análise de um conjunto de dados históricos, busca-se criar um sistema capaz de prever o volume de veículos em determinadas vias e horários, auxiliando na identificação de padrões de comportamento e na antecipação de situações de congestionamento.

Objetivo do Teste:
O teste de Shapiro-Wilk ajuda a verificar se as variáveis independentes atendem aos pressupostos de normalidade, que são importantes para certas técnicas estatísticas
O teste de Shapiro-Wilk é usado para verificar se uma amostra de dados segue uma distribuição normal.

```python
colunas = ['CarCount', 'BikeCount', 'BusCount', 'TruckCount']

for coluna in colunas:
    stat, p = stats.shapiro(df[coluna])
    print(f'Teste de Shapiro-Wilk para {coluna}:')
    print(f'  Estatística de teste: {stat}')
    print(f'  Valor p: {p}')
    if p > 0.05:
        print('  A amostra parece seguir uma distribuição normal (falha ao rejeitar H0)')
    else:
        print('  A amostra não parece seguir uma distribuição normal (rejeita H0)')
    print('-' * 30)
```

```python
variaveis = ['CarCount', 'BusCount', 'BikeCount', 'TruckCount']

# Criando a grade de subplots (2 linhas, 2 colunas)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Iterando sobre as variáveis e plotando os gráficos Q-Q
for i, variavel in enumerate(variaveis):
    row = i // 2  # Calcula a linha do subplot
    col = i % 2   # Calcula a coluna do subplot
    sm.qqplot(df[variavel], line='s', ax=axes[row, col])
    axes[row, col].set_title(f'Q-Q Plot de {variavel}')

# Ajustando o layout para evitar sobreposição de títulos
plt.tight_layout()
plt.show()
```

A ANOVA é robusta a pequenos desvios da normalidade, então se seus dados não fogem muito da linha central do grafico Q-Q, a anova ainda pode ser usada.

O teste de Levene é usado para verificar a homogeneidade das variâncias entre grupos

```python
def teste_levene_para_colunas(df, colunas, coluna_agrupamento):
  
    for coluna in colunas:
        grupos = []
        for grupo in df[coluna_agrupamento].unique():
            grupos.append(df[df[coluna_agrupamento] == grupo][coluna])

        stat, p = stats.levene(*grupos)

        print(f'Teste de Levene para {coluna}:')
        print(f'  Estatística de teste: {stat}')
        print(f'  Valor p: {p}')

        if p > 0.05:
            print('  As variâncias são homogêneas (falha ao rejeitar H0)')
        else:
            print('  As variâncias não são homogêneas (rejeita H0)')
        print('-' * 30)

colunas_independentes = ['CarCount', 'BikeCount', 'BusCount', 'TruckCount']
coluna_agrupamento = 'Traffic Situation'

teste_levene_para_colunas(df, colunas_independentes, coluna_agrupamento)
```

```python
df.head()
```

A ANOVA foi escolhida devido à sua robustez com grandes conjuntos de dados, conforme observado no gráfico Q-Q que indica desvios mínimos da normalidade. Além disso, a ANOVA é eficaz para comparar médias entre múltiplos grupos, o que é essencial para analisar o impacto da variável categórica 'situationtraffic' nas contagens de veículos. Embora o teste de Levene tenha revelado variâncias não homogêneas, estou ciente das limitações e planejando análises de resíduos e testes post-hoc robustos para validar os resultados

```python
def anova_para_colunas(df, colunas, coluna_agrupamento):
   
    for coluna in colunas:
        grupos = []
        for grupo in df[coluna_agrupamento].unique():
            grupos.append(df[df[coluna_agrupamento] == grupo][coluna])

        stat, p = stats.f_oneway(*grupos)

        print(f'ANOVA para {coluna}:')
        print(f'  Estatística de teste: {stat}')
        print(f'  Valor p: {p}')

        if p < 0.05:
            print('  Existem diferenças significativas entre os grupos (rejeita H0)')
        else:
            print('  Não existem diferenças significativas entre os grupos (falha ao rejeitar H0)')
        print('-' * 30)

colunas_independentes = ['CarCount', 'BikeCount', 'BusCount', 'TruckCount']
coluna_agrupamento = 'Traffic Situation'

anova_para_colunas(df, colunas_independentes, coluna_agrupamento)
```

 o teste de Tukey é uma escolha comum para análises post-hoc após uma ANOVA, especialmente quando você deseja comparar todas as combinações de pares de médias. No entanto, é importante lembrar que, como você identificou variâncias não homogêneas, o teste de Tukey tradicional pode não ser o mais adequado

```python
variaveis_numericas = ['CarCount', 'BusCount', 'BikeCount', 'TruckCount']
coluna_categoria = 'Traffic Situation'

for variavel in variaveis_numericas:
    # Realizando o teste de Tukey
    tukey_result = pairwise_tukeyhsd(df[variavel], df[coluna_categoria])

    # Imprimindo os resultados
    print(f"Teste de Tukey para {variavel}:\n")
    print(tukey_result)
    print("-" * 40)  # Separador para melhor visualização
```

"A análise de variância (ANOVA) revelou diferenças estatisticamente significativas nas contagens médias de carros, ônibus, bicicletas e caminhões entre as diferentes situações de tráfego analisadas. Especificamente, o teste post-hoc de Tukey demonstrou que todas as comparações de pares entre as situações de tráfego foram significativas (p < 0,05), indicando que cada situação de tráfego apresenta contagens médias de veículos distintas. Esses resultados sugerem que a composição do tráfego varia substancialmente entre as diferentes situações, com implicações potenciais para o planejamento de tráfego e a gestão de congestionamentos."

"Nossa análise mostrou que a quantidade de carros, ônibus, bicicletas e caminhões muda bastante dependendo da situação do trânsito. Descobrimos que cada tipo de situação de trânsito (como trânsito normal, lento ou intenso) tem uma quantidade diferente de veículos. Isso significa que a mistura de veículos nas ruas varia muito conforme o trânsito muda, o que pode nos ajudar a entender melhor e planejar o tráfego."

### Regressão Logística - Versão 1 - Modelo Base

- Modelo sem otimização de hiperparâmetros.
- Treinamento com dados de entrada padronizados.
- Treinamento com classes desbalanceadas na variável target.
- Avaliação com dados entrada de teste padronizados.
- Avaliação com classes desbalanceadas na variável target.

Métricas de Avaliação para Classificação de Tráfego:

Cohen Kappa Score:

Mede a concordância entre as previsões do modelo e os rótulos reais, levando em conta a chance de concordância aleatória.
Útil para lidar com classes desbalanceadas.

Recall (Sensibilidade):

Calcula a proporção de casos positivos reais que foram identificados corretamente pelo modelo.
Importante para minimizar falsos negativos (casos de congestionamento não detectados).

Precision (Precisão):

Calcula a proporção de previsões positivas que estavam corretas.
Importante para minimizar falsos positivos (previsões de congestionamento incorretas).

Matriz de Confusão:

Exibe o número de verdadeiros positivos, falsos positivos, verdadeiros negativos e falsos negativos.
Permite visualizar o desempenho do modelo em cada classe.

Classification Report:

Fornece um resumo de Precision, Recall, F1-score e suporte para cada classe.
Facilita a análise do desempenho geral do modelo.

## Modelo base: Regressão Linear

```python
# Cria o modelo
modelo_v1 = LogisticRegression()

# Treina o modelo com dados de entrada padronizados
modelo_v1.fit(X_treino_scaled, y_treino)

# Previsões com o modelo treinado
previsoes_v1 = modelo_v1.predict(X_teste_scaled)

# Matriz de confusão
print("\nMatriz de Confusão:\n")
conf_mat = confusion_matrix(y_teste, previsoes_v1)
tags = ["intenso", "nao_intenso"]
sns.heatmap(conf_mat, annot = True, xticklabels = tags, yticklabels = tags, cmap = "RdPu")
plt.show()

# Calcula métricas tradicionais

# Métricas Locais
recall_classe0_v1 = recall_score(y_teste, previsoes_v1, pos_label = 0)
recall_classe1_v1 = recall_score(y_teste, previsoes_v1, pos_label = 1)
precision_classe0_v1 = precision_score(y_teste, previsoes_v1, pos_label = 0)
precision_classe1_v1 = precision_score(y_teste, previsoes_v1, pos_label = 1)
print("Recall Classe 0 = {}".format(recall_classe0_v1))
print("Recall Classe 1 = {}".format(recall_classe1_v1))
print("Precision Classe 0 = {}".format(precision_classe0_v1))
print("Precision Classe 1 = {}".format(precision_classe1_v1))

# Métricas Globais
cohen_kappa_v1 = cohen_kappa_score(y_teste, previsoes_v1)
acc_v1 = modelo_v1.score(X_teste_scaled, y_teste)
print("\nCoeficiente Cohen kappa = {}".format(cohen_kappa_v1))
print("Acurácia = {}".format(acc_v1))

# Relatório de classificação
report_v1 = classification_report(y_teste, previsoes_v1)
print("\nRelatório de Classificação:\n\n", report_v1)
```

```python
resultados_lr = {
    'Modelo': ['Regressão Linear'],
    'Acurácia': [0.9379],
    'Cohen Kappa Score': [0.8373],
    'Recall (Classe 0)': [0.8706],
    'Recall (Classe 1)': [0.9615],
    'Precision (Classe 0)': [0.8878],
    'Precision (Classe 1)': [0.9550],
    'F1-score (Classe 0)': [0.88],
    'F1-score (Classe 1)': [0.96]
}

lr_resultados = pd.DataFrame(resultados_lr)
lr_resultados
```

**Observações:**

* O modelo de regressão linear apresentou um excelente desempenho na classificação de tráfego.
* O Cohen Kappa Score indica uma forte concordância entre as previsões e os rótulos reais.
* O modelo performou melhor na identificação de tráfego intenso (Classe 1).
